# Visualization for DLAI papers

## The quickest and best python library for nice looking graphs is seaborn.
### Otherwise you are going dash oriented.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
cols = ['cui','old_document', 'new_document', 'og_fk_score','new_fk_score','og_fk_grade','new_fk_grade', 'og_dc_score',
        'new_dc_score','og_cl_score','new_cl_score','og_ar_index','new_ar_index', 'og_smog_score', 'new_smog_score',
        'og_gf_index', 'new_gf_index']
data = pd.read_csv('pipeline_results_graded.txt', sep='|', names=cols)
data

```

```python
# Draw a nested boxplot to show bills by day and time
#iris = pd.melt(data)
#sns.stripplot(data=data, dodge=True, jitter=1.000000001)
```

```python
sns.set_style("whitegrid")
sns.set_color_codes()
```

```python
sns.distplot(data.og_fk_score, kde=True, bins=50,color='r')
#plt.axvline(x=data.og_fk_score.mean(),color='red')
sns.distplot(data.new_fk_score, kde=True, bins=50,color='g')
#plt.axvline(x=data.new_fk_score.mean(),color='green')
plt.title("GPT Flesch-Kincaid score distributions", fontsize=20, color = 'black')
plt.legend(labels=['Original','Substituted'])
plt.xlabel('FK score')
plt.ylabel('Count')
plt.xlim(-10,100)
#plt.show()
plt.savefig('/home/karl/Pictures/DLAI/GPT/fk_score_gpt.png', bbox_inches='tight')
```
