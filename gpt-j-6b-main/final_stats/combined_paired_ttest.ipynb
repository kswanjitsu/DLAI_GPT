```python
"""
Delta of model/method t-test
Compare GPT T5 and programmatic with heads up A vs B, A vs C and B vs C for the best effect
"""
```

```python
from tqdm.notebook import tqdm
import pingouin as pt
import pandas as pd
```

```python
handle_gpt = "/home/karl/PycharmProjects/DLAI_GPT/gpt-j-6b-main/final_stats/pipeline_results_graded_gpt.txt"
handle_t5 = "/home/karl/PycharmProjects/DLAI_GPT/gpt-j-6b-main/final_stats/t5_pipeline_graded_results.txt"
handle_prog = "/home/karl/PycharmProjects/DLAI_GPT/gpt-j-6b-main/final_stats/pipeline_results_graded_prog.txt"

prog_list = []
t5_list = []

for line1 in open(handle_t5):
    t5_list.append(line1.lower())

for line2 in open(handle_prog, 'r'):
    prog_list.append(line2.lower())

merge_prog_t5_list = list(zip(prog_list, t5_list))
```

```python
for i in merge_prog_t5_list[0:9]:
    print(i[0])
    print(i[1])


print(len(merge_prog_t5_list))
```

```python
gpt_list = []
comb_list = []
for line0 in tqdm(open(handle_gpt, 'r')):
    gpt_def = line0.split('|')[1].lower()
    #print(gpt_def)
    for tup in merge_prog_t5_list:
        if str(gpt_def.lower()) in str(tup[0].lower()):
            #print('lol')
            comb_list.append([tup[0],tup[1], '|'.join(line0.split('|')[1:])])

#if def_gpt in def_t5
#comb_defs = [line0, line1, line2]
#comb_list.append(comb_defs)
```

```python
for i in comb_list:
    print(i[0])
    print(i[1])
    print(i[2])
    #print(i[2])

print(len(comb_list))
```

```python
unique_list = []
for i in comb_list:
    if i not in unique_list:
        unique_list.append(i)
print(len(unique_list))
```

```python
cols = [0'old_document', 1'new_document', 2'old_fk_score',3'new_fk_score',4'old_fk_grade',5'new_fk_grade', 6'old_dc_score',
        7'new_dc_score',8'old_cl_score',9'new_cl_score',10'old_ar_index',11'new_ar_index', 12'old_smog_score', 13'new_smog_score',
        14'old_gf_index', 15'new_gf_index']
```

```python
fk_score_list = []
fk_grade_list = []
dc_grade_list = []
cl_grade_list = []
ar_grade_list = []
smog_grade_list = []
gf_grade_list = []
for i in unique_list:
    prog = i[0]
    prog_split = prog.split('|')
    prog_delta_fk_score = float(prog_split[3]) - float(prog_split[2])
    prog_delta_fk_grade = float(prog_split[5]) - float(prog_split[4])
    prog_delta_dc_grade = float(prog_split[7]) - float(prog_split[6])
    prog_delta_cl_grade = float(prog_split[9]) - float(prog_split[8])
    prog_delta_ar_grade = float(prog_split[11]) - float(prog_split[10])
    prog_delta_smog_grade = float(prog_split[13]) - float(prog_split[12])
    prog_delta_gf_grade = float(prog_split[15]) - float(prog_split[14])


    t5 = i[1]
    t5_split = t5.split('|')
    t5_delta_fk_score = float(t5_split[3]) - float(t5_split[2])
    t5_delta_fk_grade = float(t5_split[5]) - float(t5_split[4])
    t5_delta_dc_grade = float(t5_split[7]) - float(t5_split[6])
    t5_delta_cl_grade = float(t5_split[9]) - float(t5_split[8])
    t5_delta_ar_grade = float(t5_split[11]) - float(t5_split[10])
    t5_delta_smog_grade = float(t5_split[13]) - float(t5_split[12])
    t5_delta_gf_grade = float(t5_split[15]) - float(t5_split[14])

    gpt = i[2]
    gpt_split = gpt.split('|')
    gpt_delta_fk_score = float(gpt_split[3]) - float(gpt_split[2])
    gpt_delta_fk_grade = float(gpt_split[5]) - float(gpt_split[4])
    gpt_delta_dc_grade = float(gpt_split[7]) - float(gpt_split[6])
    gpt_delta_cl_grade = float(gpt_split[9]) - float(gpt_split[8])
    gpt_delta_ar_grade = float(gpt_split[11]) - float(gpt_split[10])
    gpt_delta_smog_grade = float(gpt_split[13]) - float(gpt_split[12])
    gpt_delta_gf_grade = float(gpt_split[15]) - float(gpt_split[14])

    fk_score_list.append((prog_delta_fk_score, t5_delta_fk_score, gpt_delta_fk_score))
    fk_grade_list.append((prog_delta_fk_grade, t5_delta_fk_grade, gpt_delta_gf_grade))
    dc_grade_list.append((prog_delta_dc_grade, t5_delta_dc_grade, gpt_delta_dc_grade))
    cl_grade_list.append((prog_delta_cl_grade, t5_delta_cl_grade, gpt_delta_cl_grade))
    ar_grade_list.append((prog_delta_ar_grade, t5_delta_ar_grade, gpt_delta_ar_grade))
    smog_grade_list.append((prog_delta_smog_grade, t5_delta_smog_grade, gpt_delta_smog_grade))
    gf_grade_list.append((prog_delta_gf_grade, t5_delta_gf_grade, gpt_delta_gf_grade))
```

```python
"""
Get fk score heads up
"""
```

```python
df_fk_score = pd.DataFrame(fk_score_list, columns=["prog", "t5", "gpt"])
df_fk_score
```

```python
print("Average delta prog: ", df_fk_score['prog'].mean())
print("Average delta t5: ", df_fk_score['t5'].mean())
print("Average delta gpt", df_fk_score['gpt'].mean())
```

```python
ttest_fk_score_pvt = pt.ttest(df_fk_score['prog'], df_fk_score['t5'],paired=True)
print(ttest_fk_score_pvt)
```

```python
ttest_fk_score_pvg = pt.ttest(df_fk_score['prog'], df_fk_score['gpt'],paired=True)
print(ttest_fk_score_pvg)
```

```python
ttest_fk_score_tvg = pt.ttest(df_fk_score['t5'], df_fk_score['gpt'],paired=True)
print(ttest_fk_score_tvg)
```
